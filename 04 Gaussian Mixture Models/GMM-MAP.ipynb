```python
import pandas as pd
import numpy as np

# Cnnformação dos Dados
data = pd.read_csv('clustering_data.csv' ,header = None)

X = np.matrix(data.values)

def normal(x, mi, sig):
    return (2*np.pi)**(-mi.shape[1]/2.0)*(np.linalg.det(sig))**(-1/2) * np.exp(-1/2 * (x - mi) @ np.linalg.pinv(sig) @ (x - mi).T)

def make_fig(mis, sigs, r, K, suffix):

    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib.pyplot as plt
    from scipy.stats import multivariate_normal
    import scipy.stats as st

    fig, ax = plt.subplots(ncols = 2, nrows = 1, figsize = [16,6])
    ax[0].plot(loglikelihoods)
    ax[0].set_title('Loglikelihood')

    cdic = ['blue','red','green','yellow','pink','grey', 'purple','black','cyan','orange']
    cluster = np.argmax(r, axis=1)
    colors = [cdic[i] for i in cluster]


    # create a grid of (x,y) coordinates at which to evaluate the kernels
    xlim = (-12, 12)
    ylim = (-12, 12)
    xres = 100
    yres = 100

    x = np.matrix(np.linspace(xlim[0], xlim[1], xres)).T
    y = np.matrix(np.linspace(ylim[0], ylim[1], yres)).T
    xx, yy = np.mgrid[-12:12:100j, -12:12:100j]
    xxyy = np.c_[xx.ravel(), yy.ravel()]

    for k in range(K):
        mvnormal = multivariate_normal(
            mean=np.squeeze(np.array(mis[k])), cov=sigs[k]
        )
        z = mvnormal.pdf(xxyy)
        ax[1].contour(xx, yy, z.reshape((100,100)), colors=cdic[k]);
        ax[1].plot(mis[k][0, 0], mis[k][0, 1], linestyle = '', marker = 'o', markersize = 15, color = cdic[k])

    for j in range(len(X)):
        ax[1].plot(X[j, 0], X[j, 1], linestyle = '', marker = 'o', markersize = 5, color = colors[j])
        

    ax[1].set_title(f'{suffix}')

    
    fig.savefig(f'output/{suffix}_K_{K}.png')
```

```python
def EM(K, pis, mis, sigs, epochs = 100):

    loglikelihoods = []
    
    for epoch in range(epochs):

        r = np.zeros((X.shape[0], K))

        # Passo E
        for k in range(K):
            r[:, k] = (pis[k] * np.apply_along_axis(normal, 1, X, mis[k], sigs[k])) / sum([pis[klinha] * np.apply_along_axis(normal, 1, X, mis[klinha], sigs[klinha]) for klinha in range(K)])

        # Passo M
        pis = r.sum(axis = 0)/X.shape[0]
        for k in range(K):
            mis[k] = (r[:, k].T @ X) / r[:, k].sum()
            a = np.zeros((2, 2))
            for i in range(X.shape[0]):
                a += (r[i, k] * (X[i, :] - mis[k]).T @ (X[i, :] - mis[k]))
            sigs[k] = a/r[:, k].sum()

        loglikelihood = (r @ np.log(pis)).sum() + (r[:, k] @ np.apply_along_axis(normal, 1, X, mis[k], sigs[k]).T).sum()
        loglikelihoods.append(loglikelihood)

    return pis, mis, sigs, r, loglikelihoods

def EMMAP(K, pis, mis, sigs, D, m0, S0, v0, k0, alphas, epochs = 100):

    loglikelihoods = []
    
    for epoch in range(epochs):

        r = np.zeros((X.shape[0], K))

        # Passo E
        for k in range(K):
            r[:, k] = (pis[k] * np.apply_along_axis(normal, 1, X, mis[k], sigs[k])) / sum([pis[klinha] * np.apply_along_axis(normal, 1, X, mis[klinha], sigs[klinha]) for klinha in range(K)])

        # Passo M
        pis = (r.sum(axis = 0) + alphas - 1)/(X.shape[0] - K + alphas.sum())
        for k in range(K):
            xk = r[:, k].T @ X / r[:, k].sum()
            mis[k] = (k0*m0[k] + xk * r[:, k].sum()) / (k0 + r[:, k].sum())
            a = S0[k] + (r[:, k].sum() * k0 / (r[:, k].sum() + k0) * (xk - m0[k]).T @ (xk - m0[k]))
            for i in range(X.shape[0]):
                a += (r[i, k] * (X[i, :] - xk).T @ (X[i, :] - xk))                

            sigs[k] = a/(v0 + D + 2 + r[:, k].sum())

        loglikelihood = (r @ np.log(pis)).sum() + (r[:, k] @ np.apply_along_axis(normal, 1, X, mis[k], sigs[k]).T).sum()
        loglikelihoods.append(loglikelihood)

    return pis, mis, sigs, r, loglikelihoods
```

```python
# K-means
mis = [
    np.matrix([[-8,-9]]),
    np.matrix([[-11,-11]]),
    np.matrix([[7,10]]),
    np.matrix([[6,1]]),
    np.matrix([[4,0]]),
    np.matrix([[-10,7]]),
    np.matrix([[10,-5]]),
    np.matrix([[10,5]]),
    np.matrix([[13,-4]]),
    np.matrix([[0,0]]),
]
K = len(mis)
pis = [1/K]*K
sigs = [np.matrix(np.eye(X.shape[1])*4)]*K

pis, mis, sigs, r, loglikelihoods = EM(K, pis, mis, sigs, epochs = 10)
make_fig(mis, sigs, r, K, 'Normal')
```

```python
# MAP
D = X.shape[1]
m0 = [X.mean(axis = 0) for mi in mis]
S0 = [1/K*np.diagflat(np.diag(np.cov(X.T))) for sig in sigs]
v0 = D + 2;
k0 = 0.01
alphas = np.array([2]*K)

pis, mis, sigs, r, loglikelihoods = EMMAP(K, pis, mis, sigs, D, m0, S0, v0, k0, alphas, epochs = 20)
make_fig(mis, sigs, r, K, 'MAP')
```
