```python
import matplotlib.pyplot as plt
def plot_wind_speed(data):
    cols = list(data.columns)
    cols = [i for i in cols if 'ws' in i]
    fig, ax = plt.subplots(nrows = 3, ncols = 5, figsize = [16, 7])
    for i in range(len(cols)):
        if 'ws' in cols[i]:
            ax[i // 5, i % 5].scatter(data[cols[i]], data['power'], s = 2)
            ax[i // 5, i % 5].set_yticks([])
            ax[i // 5, i % 5].set_xticks([])
    ax[1, 0].set_ylabel('Power')
    ax[1, 0].set_yticks([0, 1000, 2000])
    ax[2, 2].set_xlabel('Wind Speed')
    ax[2, 2].set_xticks([0, 5, 10, 15, 20])
    ax[2, 4].set_yticks([])
    ax[2, 4].set_xticks([])
    ax[2, 4].scatter(data[cols].mean(axis = 1), data['power'], s = 2, color = '#bdbdbd')
    ax[2, 4].legend(['mean'])
    
plot_wind_speed(data_train)
```

```python
import pandas as pd
def adjust_dataset(data):

    data = data.copy()
    
    cols = list(data.columns)

    cols.remove('power')
    cols.remove('limited')
    
    X = data[cols].values
    Y = (data['power'].values)[..., None]
    Z = (data['limited'].values)[..., None]

    return X, Y, Z
```

```python
X_train, Y_train, _ = adjust_dataset(data_train)
X_test, Y_test, Z_test = adjust_dataset(data_test)
```

# Bayesian Linear Regression

O primeiro método utilizado para classificação de pontos limitados a partir dos atributos medidos nas torres anemmétrica foi o algoritmo de regresão linear bayesiana. Como conhecimento a priori, foi utilizado o algoritmo OLS para a média, e matriz identidade para covariância. FOram feitos dois testes: um utilizando grau polinomial 5 e outro utilizando grau polinomial 5 em conjunto com transformação sigmoidal nos dados.

```python
class BayesLinReg:

    def __init__(self, pol = None, sig = False):
        self.m0 = None
        self.S0 = None
        self.sig0 = None
        self.pol=pol
        self.sig=sig

    def transform(self, X, pol = None):

        if self.sig:
            X = 1 / (1 + np.exp(-X))
        
        X2 = X.copy()
        
        if pol:
            for i in range(pol-1):
                X = np.hstack((X, X2 ** i+2))
        X = np.hstack((np.ones((len(X), 1)), X))

        return X
    
    def fit(self, X, Y, m0 = None, S0 = None, sig0 = None):
        
        X = self.transform(X, pol=self.pol)
        
        # Com dimensão
        m0 = np.linalg.pinv(X.T @ X) @ X.T @ Y
        S0 = np.eye((X.shape[1]))
        sig0 = 100 if not self.sig else 0.01

        I = np.eye(X.shape[1])
                    
        m0 = m0 + np.linalg.pinv(S0 @ X.T @ X + sig0 * I) @ S0 @ X.T @ (Y - X @ m0)
        S0 = S0 - np.linalg.pinv(S0 @ X.T @ X + sig0 * I) @ S0 @ X.T @ X @ S0

        self.m0 = m0
        self.S0 = S0
        self.sig0 = sig0
    
    def predict(self, X):
        
        X = self.transform(X, pol = self.pol)
        Med = X @ self.m0
        Cov = X @ self.S0 @ X.T + self.sig0 * np.eye(len(X))
        
        return Med, Cov
```

```python
def plot_model(Y, Y_h, Y_tol):
    import matplotlib.pyplot as plt
    plt.figure(figsize = [16, 7])
    plt.plot(Y)
    plt.plot(Y_h)
    plt.plot(Y_h + 2*Y_tol)
    plt.plot(Y_h - 2*Y_tol)
```

```python
def get_metrics(Z_hat, Z_test):
    from sklearn.metrics import confusion_matrix
    confusion = confusion_matrix(Z_hat, Z_test)
    accuracy = np.diag(confusion).sum()/confusion.sum()
    precision = confusion[0,0]/confusion[:, 0].sum()
    recall = confusion[0,0]/confusion[0, :].sum()
    beta = 2
    F1 = (1 + beta ** 2) * precision * recall / (beta ** 2 * precision + recall)
    return f'accuracy {accuracy*100:.2f}% precision {precision*100:.2f}% recall {recall*100:.2f}% F-beta {F1*100:.2f}%'
```

# Common Gaussian Process

Foi usada a função de Kernel RBF, com Likelihood gaussiana. Devido à dificuldade computacional de usar uma GP com todas as entradas, foi utilizados o método de K-média para separar 200 entradas a partir do conjuntos de dados original.

```python
import pandas as pd
import numpy as np
from scipy.linalg import cho_factor, solve_triangular
from scipy.optimize import minimize
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans as km
from sklearn.covariance import EllipticEnvelope

# Definição da função logverossimililhançamarginal
def rbf(X, XJ, sig2f, wd2):
    return sig2f * np.exp (
        -1/2 * ((X[:, None] - XJ[None, :]) ** 2 * wd2).sum(axis = 2)
    )

def logvermarginal(x, X, Y):

    N = X.shape[0]
    I = np.eye(N)
    sig2f, sig2y, wd2 = x[0], x[1], x[2:]
    K = rbf(X, X, sig2f, wd2)
    Ky = K + sig2y * I
    L, _ = cho_factor(Ky, lower = True)
    alpha = solve_triangular(L, Y)
    elbo = -float(np.log(np.diag(L)).sum() + 1/2 * alpha.T @ alpha + N/2 * np.log(2*np.pi))

    return -elbo

class GPR:
    
    def __init__(self, X, Y, sig2f = None, wd2 = None, sig2y = None):
        # Definição da priori


        AUX = np.hstack((X.copy(), Y.copy()))
        AUX2 = km(n_clusters = 200).fit(AUX)
        AUX2 = AUX2.cluster_centers_

        self.X = AUX2[:, :-1]
        self.Y = AUX2[:, -1][..., None]
        self.N = self.X.shape[0]

        self.sig2f = sig2f if sig2f else 1
        self.wd2 = wd2 if isinstance(wd2, np.ndarray) else 2 / (X.max(axis = 0) - X.min(axis = 0)) ** 2
        self.sig2y = sig2y if sig2y else 0.1*self.sig2f
        self.I = np.eye(self.N)

    def fit(self, **kwargs):

        X = self.X.copy()
        Y = self.Y.copy()
        x0 = np.array([self.sig2f, self.sig2y] +  list(self.wd2))

        # Minimização utilizando o scipy (algoritmo L-BFGS-B)
        m = minimize(
            logvermarginal, 
            x0, 
            args = (X, Y), 
            bounds = [(0.0001, None)]*len(x0), 
            method = 'L-BFGS-B', 
            options = kwargs,
        )
        x = m.x
        self.sig2f, self.sig2y, self.wd2 = x[0], x[1], x[2:]

    def predict(self, X_):

        # Etapa de Predição
        K = rbf(self.X, self.X, sig2f, wd2)
        K_ = rbf(self.X, X_, sig2f, wd2)
        K__ = rbf(X_, X_, sig2f, wd2)
        Ky = K + sig2y*self.I
        L = np.linalg.cholesky(Ky)

        v = solve_triangular(L, K_, lower = True)
        u = solve_triangular(L, self.Y, lower = True)

        Y_mean = v.T @ u
        Y_cov = K__ - v.T @ v
        
        return Y_mean, Y_cov
```

```python
# Aproximação inicial dos hiperparâmetros
model = GPR(xscaler.transform(X_train), yscaler.transform(Y_train))
model.fit()
sig2f, sig2y, wd2 = model.sig2f, model.sig2y, model.wd2
```

# Stochastic Variational Gaussian Processes

Foi utilizada aqui uma abordagem por processos gaussianos utilizando o algoritmo variacional estocástico. A indução foi setada inicialmente em 400 pontos, pegos através do método K-Médias e filtrados utilizando um envelope elíptico, resultando num conjunto de indução com um número menor de pontos. Foi utilizada uma verossimilhança gaussiana com função de Kernel Matern32.

```python
import tensorflow as tf
import gpflow as gpf
import json 
import numpy as np
from sklearn.cluster import KMeans as km
from sklearn.covariance import EllipticEnvelope

class SVGP:
    
    def __init__(self, X, Y, inducing_size = 400):

        # Definindo o conjunto de indução
        M = inducing_size
        N = X.shape[0]
        self.X = X.copy()
        self.Y = Y.copy()
        self.M = M
        self.N = N
        
        Z = km(n_clusters=M).fit(X)
        Z = np.matrix(Z.cluster_centers_)

        cov = EllipticEnvelope(random_state=0).fit(Z)
        outliers = (cov.predict(Z) + 1).astype(bool)
        Z = Z[outliers]
        
        self.Z = Z

        # %% Model
        self.likelihood = gpf.likelihoods.Gaussian(variance=1.0)

        self.kernel = gpf.kernels.Matern32(
         variance=1.0,
         lengthscales=[1.0]*X.shape[1]
        )

        self.model = gpf.models.SVGP(
            kernel=self.kernel,
            likelihood=self.likelihood,
            inducing_variable=self.Z,
            num_data=N
        )

    def fit(self, epochs = 100, learning_rate = 0.01):

        N = self.X.shape[0]
        data_tuple = (self.X.copy(), self.Y.copy())
        elbo = tf.function(self.model.elbo)
        tensor_data = tuple(map(tf.convert_to_tensor, data_tuple))
        elbo(tensor_data)  # run it once to trace & compile

        minibatch_size = 100
        train_dataset = tf.data.Dataset.from_tensor_slices(data_tuple).repeat().shuffle(N)
        train_iter = iter(train_dataset.batch(minibatch_size))
        ground_truth = elbo(tensor_data).numpy()

        # We turn off training for inducing point locations
        gpf.set_trainable(self.model.inducing_variable, False)
        
        def run_adam(model, iterations):
            # Create an Adam Optimizer action
            logf = []
            train_iter = iter(train_dataset.batch(minibatch_size))
            training_loss = self.model.training_loss_closure(train_iter, compile=True)
            optimizer = tf.optimizers.Adam(learning_rate = learning_rate)

            @tf.function
            def optimization_step():
                optimizer.minimize(training_loss, self.model.trainable_variables)

            for step in range(iterations):
                optimization_step()
                if step % 10 == 0:
                    elbo = -training_loss().numpy()
                    logf.append(elbo)
            return logf

        from gpflow.ci_utils import ci_niter
        maxiter = ci_niter(epochs)
        logf = run_adam(self.model, maxiter)
        
        #return logf
    
    def predict(self, X):
        # Etapa de Predição
        Y_mean, Y_cov = self.model.predict_f(X.copy())
        Y_mean = Y_mean.numpy()
        Y_cov = Y_cov.numpy()
        
        return Y_mean, Y_cov
```
